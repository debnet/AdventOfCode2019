```python
def deal_into_new_stack(cards):
    return cards[::-1]

def cut_cards(cards, cut):
    return cards[cut:] + cards[0:cut]

def deal_with_increment(cards, increment):
    deck = [0] * len(cards)
    for i in range(len(cards)):
        deck[(i * increment)% len(cards)] = cards[i]
    return deck

def solve(actions, cards=10007, search=2019):
    cards = list(range(cards))
    for action in actions.splitlines():
        *_, param = action.split()
        if 'new' in action:
            cards = deal_into_new_stack(cards)
        elif 'cut' in action:
            cards = cut_cards(cards, int(param))
        elif 'with' in action:
            cards = deal_with_increment(cards, int(param))
    return cards.index(search)
```
