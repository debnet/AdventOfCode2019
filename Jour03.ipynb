```python
wire1 = [(w[0], int(w[1:])) for w in wire1.split(',')]
wire2 = [(w[0], int(w[1:])) for w in wire2.split(',')]
```

```python
def to_path(wire):
    px, py, opx, opy = 0, 0, 0, 0
    path = []
    for direction, length in wire:
        opx, opy = px, py
        if direction == 'R':
            px += length
        elif direction == 'L':
            px -= length
        elif direction == 'U':
            py += length
        elif direction == 'D':
            py -= length
        path.append((px, py))
    return path
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib.path import Path
import matplotlib.patches as patches

verts1, codes1 = [(0, 0)], [Path.MOVETO]
for point in to_path(wire1):
    verts1.append(point)
    codes1.append(Path.LINETO)
path1 = Path(verts1, codes1)

verts2, codes2 = [(0, 0)], [Path.MOVETO]
for point in to_path(wire2):
    verts2.append(point)
    codes2.append(Path.LINETO)
path2 = Path(verts2, codes2)

fig, ax = plt.subplots()
patch1 = patches.PathPatch(path1, edgecolor='red', facecolor='none', lw=1)
patch2 = patches.PathPatch(path2, edgecolor='blue', facecolor='none', lw=1)
ax.add_patch(patch1)
ax.add_patch(patch2)
ax.autoscale()
plt.grid()
plt.show()
```

```python
from shapely.geometry import LineString

line1 = LineString([(0, 0)] + to_path(wire1))
line2 = LineString([(0, 0)] + to_path(wire2))

def get_intersections(line1, line2):
    points = []
    for point in line1.intersection(line2):
        if point.x == point.y == 0:
            continue
        points.append(point)
    return points

intersections = get_intersections(line1, line2)
len(intersections)
```
